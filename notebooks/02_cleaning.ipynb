# Final Data Cleaning Summary

## Objective
Prepare the raw banking churn dataset for analysis by identifying and resolving data quality issues, standardizing formats, and exporting a clean version for downstream EDA, statistical analysis, and dashboarding.

## Dataset Overview
- Original Shape: **28,382 rows × 21 columns**
- Source File: `data/raw/Banking_churn_prediction.csv`

## Cleaning Steps Performed

### 1. Initial Data Audit
Performed:
- Dataset shape review
- Column datatype inspection
- Descriptive statistics
- Missing value assessment
- Duplicate row check

### 2. Duplicate Records
- Checked duplicate rows.
- No duplicate rows found / duplicates removed if detected.

### 3. Datatype Corrections
- Converted `last_transaction` from object to datetime format using safe parsing.

### 4. Missing Value Treatment
Handled null values as follows:

- `gender` → filled using mode
- `occupation` → filled using mode
- `dependents` → filled using median
- `city` → filled using mode

### 5. Date Column Handling
- `last_transaction` conversion produced **3,223 missing values (NaT)** due to blank/invalid dates.
- Missing dates were retained for future transaction-recency analysis.

### 6. Outlier / Anomaly Review

#### Age
- Minimum age observed: 1
- Ages below 18 were retained, as junior/minor banking accounts may exist.

#### Dependents
- Unrealistic values (>10) detected.
- Extreme values were replaced using median dependents value.

### 7. Format Standardization
- Converted `dependents` to integer datatype.
- Converted `city` to integer datatype.

## Final Null Value Status
- All columns cleaned successfully.
- Only `last_transaction` retains missing values intentionally.
